# Set-Up

## - Import Package

```python
# Import package
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

## - Insert Image

```python
Image = cv2.imread('/content/smallGirl.png',0)
cv2_imshow(Image)
print('Image Size: ', Image.shape)
```

## - Concolution Function

```python
def Convolution(Image, Kernel):
  m, n = Kernel.shape
  assert ((m<=Image.shape[0]) or (n<=Image.shape[1])), 'Error: Kernel is bigger than Image size'
  col_sub, col_add = m//2, -(-m//2)
  row_sub, row_add = n//2, -(-n//2)

  Output_Image = np.zeros(Image.shape)
  for i in range(col_sub,Image.shape[0]-col_sub):
    for j in range(row_sub,Image.shape[1]-row_sub):
      mat = np.multiply(Kernel,Image[i-col_sub:i+col_add,j-row_sub:j+row_add]);
      Output_Image[i,j] = (abs(np.sum(mat)));
  return Output_Image
```

# Edge Detection Kernels

##  - Sobel

```python
# Kernel: Sobel - Vertical Edge
sobel_1 = np.array([[-1,0,1], [-2,0,2],[-1,0,1]])
```

```python
Output_Image = Convolution(Image, sobel_1)
cv2_imshow(Output_Image)
```

```python
# Kernel: Sobel - Horzintal Edge
sobel_2 = np.array([[-1, -2, -1],[0, 0, 0], [1, 2, 1]])
```

```python
Output_Image = Convolution(Image, sobel_2)
cv2_imshow(Output_Image)
```

## - Prewitt

```python
# Kernel: Prewitt - Vertical Edge
prewitt_1 = np.array([[-1,0,1], [-1,0,1],[-1,0,1]])
```

```python
Output_Image = Convolution(Image, prewitt_1)
cv2_imshow(Output_Image)
```

```python
# Kernel: Prewitt - Horzintal Edge
prewitt_2 = np.array([[-1, -1, -1],[0, 0, 0], [1, 1, 1]])
```

```python
Output_Image = Convolution(Image, prewitt_2)
cv2_imshow(Output_Image)
```

## - Robert

```python
# Kernel: Robert - Vertical Edge
robert_1 = np.array([[1, 0], [0, -1]])
```

```python
Output_Image = Convolution(Image, robert_1)
cv2_imshow(Output_Image)
```

```python
# Kernel: Robert - Horzintal Edge
robert_2 = np.array([[0, 1], [-1, 0]])
```

```python
Output_Image = Convolution(Image, robert_2)
cv2_imshow(Output_Image)
```
